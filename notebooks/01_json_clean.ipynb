A csv fájlhoz hasonlóan, a műfajok itt is felsorolásként vannak megadva.

```python
bad_values = ["", "null", "n/a", "none", "na", "[]"]

df_json.replace(["", "[]", "null", "N/A", "None", "NaN"], pd.NA, inplace=True)
```

```python
df_json.isnull().sum()
```

```python
missing_pct = df_json.isnull().sum()/len(df_json)

kept_cols = missing_pct[missing_pct < 0.5].index

df_json = df_json[kept_cols]
df_json.info()
```

Eldobtam az oszlopokat, ahol az adatok legalább fele hiányos volt. Feltöltöm a szöveges null adatokat, de a numerikusokat meghagyom.

```python
object_cols = df_json.select_dtypes(include=['object']).columns.tolist()

df_json[object_cols] =df_json[object_cols].fillna("Unknown")

df_json.isnull().sum()
```

```python
def fix_empty_lists(val):
    bad_values = ["", "[]", "null", "n/a", "none", "na", "nan"]
    
    if isinstance(val, list):
        cleaned_list = [item for item in val if str(item).strip().lower() not in bad_values]
        
        if len(cleaned_list) == 0:
            return ["Unknown"]
        
        return cleaned_list
        
    return val

list_columns = ['genres', 'developers', 'publishers']

for col in list_columns:
    df_json[col] = df_json[col].apply(fix_empty_lists)

df_json[['genres', 'developers']].head(15)
```

A numerikus értékeknél sincsenek végül null-értékek, valószínűleg azért, mert az alap datasetben 0-ként voltak reprezentálva.

Az első oszlop most indexként van jelen, de a csv fájlban AppID néven van.

```python
df_json.index.name = "appID"
df_json.reset_index(inplace=True)
```

```python
# dim_games tábla
dim_games = df_json[["appID", "name", "required_age", "about_the_game"]].rename(columns={"appID": "game_id"})
dim_games.head()
```

```python
df_json["developers"].head(20)
```

```python
# dim_developers tábla
# mivel van olyan játék aminek több fejlesztője van, az elsőt tartom meg a felsoroltak közül
df_json["developer_name"] = df_json["developers"].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else x)

dim_developers = df_json[["developer_name"]].drop_duplicates().dropna()

dim_developers["developer_id"] = range(1, len(dim_developers) + 1)

dim_developers.head()
```

```python
df_json["publishers"].head(20)
```

```python
# dim_publishers tábla
# mivel van olyan játék aminek több kiadója van, az elsőt tartom meg a felsoroltak közül
df_json["publisher_name"] = df_json["publishers"].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else x)

dim_publishers = df_json[["publisher_name"]].drop_duplicates().dropna()

dim_publishers["publisher_id"] = range(1, len(dim_publishers) + 1)

dim_publishers.head()
```

--

```python
# dim_date tábla
df_json["release_date_parsed"] = pd.to_datetime(df_json["release_date"], errors='coerce')

unique_dates = df_json[["release_date_parsed"]].dropna().drop_duplicates()
dim_date = unique_dates.rename(columns={"release_date_parsed" : "release_date"})
dim_date["date_id"] = range(1, len(dim_date) + 1)
dim_date["release_year"] = dim_date["release_date"].dt.year
dim_date["release_month"] = dim_date["release_date"].dt.month
dim_date["release_day"] = dim_date["release_date"].dt.day

dim_date.head()
```

--

Első ötletnek minden műfaj első entry-jét tartottam volna meg, de mivel névsorban vannak felsorolva, ez nem a legjellemzőbb minden játékra, és az alfabetikusan hamarabb levő szavak többször fordulnának elő mint amik hátrébb vannak. Ezért inkább a "bridge table" megoldást alkalmazom.

```python
# minden műfajt szétválasztok, és megszűntetem a duplikációkat
all_genres = df_json["genres"].explode().unique()

dim_genres = pd.DataFrame(all_genres, columns=["genre_name"]).dropna()
dim_genres = dim_genres.sort_values("genre_name").reset_index(drop=True)

# 1-től kezdődő indexek
dim_genres["genre_id"] = dim_genres.index + 1

dim_genres.head(50)
```

```python
# bridge_game_genres tábla
temp = df_json[["appID", "genres"]].copy()
temp = temp.explode("genres")
temp = temp.rename(columns={"genres": "genre_name"})
temp = temp.rename(columns={"appID" : "game_id"})

bridge_game_genres = temp.merge(dim_genres, on="genre_name", how="left")
bridge_game_genres = bridge_game_genres[["game_id", "genre_id"]].dropna().astype(int)

bridge_game_genres.head(10)
```

```python
numeric_cols = df_json.select_dtypes(include=['number']).columns.tolist()

print(f"There are {len(numeric_cols)} numerical columns:")
print(numeric_cols)
```

```python
df_json[['achievements', 'positive', 'negative', 'average_playtime_forever','discount']].head(50)

# 'achievements', 'positive', 'negative', 'average_playtime_forever', 'average_playtime_2weeks', 'median_playtime_forever', 'median_playtime_2weeks', 'discount', 'peak_ccu'
```

```python
# fact_game_statistics tábla
fact_temp = df_json.merge(dim_developers, left_on="developer_name", right_on="developer_name", how="left")
fact_temp = fact_temp.merge(dim_publishers, left_on="publisher_name", right_on="publisher_name", how="left")
fact_temp = fact_temp.merge(dim_date, left_on="release_date_parsed", right_on="release_date", how="left")

fact_game_statistics = fact_temp[[
    "appID",
    "developer_id",
    "publisher_id",
    "date_id",
    "price",
    "achievements",
    "average_playtime_forever",
    "discount", 
    "peak_ccu"
]]

fact_game_statistics = fact_game_statistics.rename(columns={"appID": "game_id"})
fact_game_statistics["discount"] = fact_game_statistics["discount"] * 0.01

fact_game_statistics.head()
```

```python
dim_games.to_csv("../data/processed/dim_games.csv", index=False)
dim_developers.to_csv("../data/processed/dim_developers.csv", index=False)
dim_publishers.to_csv("../data/processed/dim_publishers.csv", index=False)
dim_date.to_csv("../data/processed/dim_date.csv", index=False)
dim_genres.to_csv("../data/processed/dim_genres.csv", index=False)
bridge_game_genres.to_csv("../data/processed/bridge_game_genres.csv", index=False)
fact_game_statistics.to_csv("../data/processed/fact_game_statistics.csv", index=False)
```
